```python
import csv
from collections import Counter

# ---------------------------------------------------------
# 1. Austausch des Kommas als Trennzeichen in der CSV-Datei durch Semikolon
# ---------------------------------------------------------

pfad_original_daten = "daten/gruen_opendata.csv"
pfad_semikolon_daten = "daten/gruen_opendata_semicolon.csv"

with open(pfad_original_daten, "r", encoding="utf-8") as original_daten, \
     open(pfad_semikolon_daten, "w", encoding="utf-8", newline="") as ausgabedatei:

    reader = csv.reader(original_daten, delimiter=",")
    writer = csv.writer(ausgabedatei, delimiter=";")

    for row in reader:
        writer.writerow(row)
```

2. Welche Baumarten gibt es und wie oft kommen sie vor?
(inkl. Rückmeldung bei fehlerhaften Zeilen)

```python
# ---------------------------------------------------------
# 2. Welche Baumarten gibt es und wie oft kommen sie vor?
#    (inkl. Rückmeldung bei fehlerhaften Zeilen)
# ---------------------------------------------------------

baumarten_zaehler = Counter()
fehlerhafte_zeilen = []   # sammelt Zeilen, die nicht genug Spalten haben (ist hier nicht der Fall gewesen)

with open(pfad_semikolon_daten, "r", encoding="utf-8") as original_daten:
    csv_leser = csv.reader(original_daten, delimiter=";")

    for zeilennummer, zeile in enumerate(csv_leser, start=1):
        if len(zeile) < 3:
            fehlerhafte_zeilen.append((zeilennummer, zeile))
            continue

        baumart = zeile[2].strip()    #strip, falls z.B.Leerzeichen vorkommen
        if baumart:
            baumarten_zaehler[baumart] += 1

anzahl_baumarten = len(baumarten_zaehler)

# ---------------------------------------------------------
# Ausgabe der Ergebnisse
# ---------------------------------------------------------

print("Anzahl unterschiedlicher Baumarten:", anzahl_baumarten)

print("\nHäufigkeit pro Baumart:")
for art, anzahl in baumarten_zaehler.most_common():
    print(f" - {art}: {anzahl}")
```

4. Rückmeldung zu fehlerhaften Zeilen

```python
# ---------------------------------------------------------
# 4. Rückmeldung zu fehlerhaften Zeilen
# ---------------------------------------------------------

if fehlerhafte_zeilen:
    print("\nAchtung: Unerwartete Zeilen gefunden:")
    for nr, inhalt in fehlerhafte_zeilen:
        print(f" - Zeile {nr}: {inhalt}")
else:
    print("\nAlle Zeilen hatten die erwartete Spaltenanzahl.")
```

Alternative zu most_common

```python
for art, anzahl in sorted(baumarten_zaehler.items(), key=lambda x: x[1], reverse=True):
    #print(art, anzahl)
    print(f" - {art}: {anzahl}")
```

```python
# und nach Alphabet:
for art, anzahl in sorted(baumarten_zaehler.items(), key=lambda x: x[0], reverse=False):
    #print(art, anzahl)
    print(f" - {art}: {anzahl}")
```
